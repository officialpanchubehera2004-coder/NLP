# Bigrams, Trigrams,Ngrams

```python

from nltk.util import bigrams,trigrams,ngrams
```

```python
string = 'generative ai & agentic ai is automate the industry & more focus on every domain'
quotes_tokens=nltk.word_tokenize(string)
quotes_tokens
```

```python
string
```

```python
len(quotes_tokens)
```

```python
qoutes_bigrams = list(nltk.bigrams(quotes_tokens))
qoutes_bigrams
```

```python
qoutes_trigram =list(nltk.trigrams(quotes_tokens))
qoutes_trigram 
```

```python
qoutes_nrigram =list(nltk.ngrams(quotes_tokens,4))
qoutes_nrigram
```

```python
from nltk.stem import PorterStemmer
pst = PorterStemmer()
```

```python
pst.stem('affection')
```

```python
pst.stem('playing')
```

```python
pst.stem('maximum')
```

```python
words_to_stem = ['give','giving','given','thinking','loving','maximum']

for words in words_to_stem:
    print(words+':'+ pst.stem(words))
```

```python
from nltk.stem import LancasterStemmer
pst = LancasterStemmer()

for words in words_to_stem:
    print(words+':'+pst.stem(words))
```

```python
from nltk.stem import SnowballStemmer
sbst = SnowballStemmer('english')   #Choose a language


for words in words_to_stem:
    print(words+':'+ sbst.stem(words))
```

```python
from nltk.tokenize import word_tokenize
```

```python
sent = 'my name is panchu'
sent_tokens = word_tokenize(sent)
sent_tokens
```

```python
for token in sent_tokens:
    print(nltk.pos_tag([token]))
```

```python
sent2 = 'john is eating a delicious cake'
sent2_tokens = word_tokenize(sent2)

for token in sent2_tokens:
    print(nltk.pos_tag([token]))
```

```python
from nltk import ne_chunk
```

```python
ne_sent = 'Rohit sharma works at hyderabad in google'
ne_tokens = word_tokenize(ne_sent)
ne_tokens
```

```python
ne_tags = nltk.pos_tag(ne_tokens)
ne_tags
```

```python
ne_ner = ne_chunk(ne_tags)
print(ne_ner)
```
